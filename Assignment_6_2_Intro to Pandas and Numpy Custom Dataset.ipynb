```python
# Choose a column with a numerical value, change its type to int64 and store in a new column.
# since there are no numerical non-int64 values, i'll change capital gain into float64

census_df['CapitalGain'] = census_df['capital-gain'].astype('float64')
census_df.dtypes
```

```python
# Choose a column with a non-numerical value (that can act as a category), create a new column that
# contains the mapped values of the categorical column from 0 to n-1.

# checking values
workclass_values = census_df['workclass'].unique()
workclass_values
```

```python
workclass_mapping = {value: index for index, value in enumerate(workclass_values)}
workclass_mapping
```

```python
census_df['workclass-num'] = census_df['workclass'].map(workclass_mapping)
census_df
```

```python
# was showing false below but then realised the csv had unnecessary space formatting

unique_values = census_df['attributes'].unique()
unique_values
```

```python
# Create a column that would have the value YES or NO depending on a conditional statement you will create based on values in the dataset.

census_df['HighIncome'] = census_df['attributes'] == ' >50K'
census_df
```
